```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import random
```

```python
class NumberGuessingGameIPywidgets:
    def __init__(self):
        self.secret_number = random.randint(1, 100)
        self.guesses_taken = 0

        # Widgets for UI
        self.instruction_label = widgets.Label(value="Guess a number between 1 and 100:")
        self.guess_input = widgets.IntText(
            value=50,
            description='Your Guess:',
            disabled=False
        )
        self.guess_button = widgets.Button(description="Guess")
        self.message_output = widgets.Output()
        self.new_game_button = widgets.Button(description="New Game")

        # Arrange widgets
        self.ui = widgets.VBox([
            self.instruction_label,
            self.guess_input,
            self.guess_button,
            self.message_output,
            self.new_game_button
        ])

        # Event handlers
        self.guess_button.on_click(self._check_guess)
        self.new_game_button.on_click(self._reset_game)

        # Initial state
        self.new_game_button.disabled = True # Disable until game ends

    def _check_guess(self, b):
        with self.message_output:
            clear_output()
            try:
                guess = self.guess_input.value
                self.guesses_taken += 1

                if guess < 1 or guess > 100:
                    print("Please enter a number between 1 and 100.")
                elif guess < self.secret_number:
                    print("Too low! Try again.")
                elif guess > self.secret_number:
                    print("Too high! Try again.")
                else:
                    print(f"Congratulations! You guessed the number in {self.guesses_taken} guesses!")
                    self.guess_button.disabled = True
                    self.new_game_button.disabled = False
            except Exception:
                print("Invalid input. Please enter a number.")
            # Clear the input box after each guess
            self.guess_input.value = 0 # Or set to a default, or leave for next input

    def _reset_game(self, b):
        with self.message_output:
            clear_output()
            self.secret_number = random.randint(1, 100)
            self.guesses_taken = 0
            print("Guess a number between 1 and 100:")
            self.guess_input.value = 50
            self.guess_button.disabled = False
            self.new_game_button.disabled = True

    def start_game(self):
        display(self.ui)
        with self.message_output:
            print("Guess a number between 1 and 100:")
```

```python
game_ipy = NumberGuessingGameIPywidgets()
game_ipy.start_game()
```

```python
import tkinter as tk
from tkinter import messagebox
import random
```

```python
class NumberGuessingGame:
    def __init__(self, master):
        self.master = master
        master.title("Number Guessing Game")

        self.secret_number = random.randint(1, 100)  # Generate a random number between 1 and 100
        self.guesses_taken = 0

        self.label = tk.Label(master, text="Guess a number between 1 and 100:")
        self.label.pack(pady=10)

        self.entry = tk.Entry(master)
        self.entry.pack(pady=5)

        self.guess_button = tk.Button(master, text="Guess", command=self.check_guess)
        self.guess_button.pack(pady=5)

        self.message_label = tk.Label(master, text="")
        self.message_label.pack(pady=10)

        self.new_game_button = tk.Button(master, text="New Game", command=self.reset_game)
        self.new_game_button.pack(pady=5)
        self.new_game_button.config(state=tk.DISABLED) # Disable until game ends

    def check_guess(self):
        try:
            guess = int(self.entry.get())
            self.guesses_taken += 1

            if guess < 1 or guess > 100:
                self.message_label.config(text="Please enter a number between 1 and 100.", fg="orange")
            elif guess < self.secret_number:
                self.message_label.config(text="Too low! Try again.", fg="red")
            elif guess > self.secret_number:
                self.message_label.config(text="Too high! Try again.", fg="red")
            else:
                self.message_label.config(text=f"Congratulations! You guessed the number in {self.guesses_taken} guesses!", fg="green")
                messagebox.showinfo("Game Over", f"You guessed the number in {self.guesses_taken} guesses!")
                self.guess_button.config(state=tk.DISABLED)
                self.new_game_button.config(state=tk.NORMAL)
        except ValueError:
            self.message_label.config(text="Invalid input. Please enter a number.", fg="red")
        self.entry.delete(0, tk.END) # Clear the entry box

    def reset_game(self):
        self.secret_number = random.randint(1, 100)
        self.guesses_taken = 0
        self.message_label.config(text="Guess a number between 1 and 100:", fg="black")
        self.entry.delete(0, tk.END)
        self.guess_button.config(state=tk.NORMAL)
        self.new_game_button.config(state=tk.DISABLED)
```
